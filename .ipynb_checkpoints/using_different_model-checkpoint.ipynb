```python
final_df=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_df.head()
```

```python
final_df=pd.get_dummies(final_df,drop_first=1)
```

```python
final_df.columns
```

```python
final_df['Current_year']=2022
```

```python
final_df['Used_years']=final_df['Current_year']-final_df['Year']
```

```python
final_df.head()
```

```python
final_df.drop(['Year'],axis=1,inplace=True)
final_df.drop(['Current_year'],axis=1,inplace=True)
```

```python
final_df.columns
```

```python
x=final_df['Used_years']
y=final_df['Selling_Price']
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("Scatterplot of used years vs selling price")
plt.xlable("Used years")
plt.ylable("Selling price")
```
